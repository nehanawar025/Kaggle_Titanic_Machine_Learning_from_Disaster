### To use the Sex column effectively I mapped the **Female: 0 and Male: 1**

```python
df['Sex'] = df.Sex.map({'female': 0, 'male': 1})

df.head()
```

## Correlation with heatmap

```python
# df.select_dtypes(include=['number']).corr() is used to get the numeric columns

sns.heatmap(df.select_dtypes(include = ['number']).corr(), annot=True, cmap="YlGnBu", vmax=1, vmin=-1)
plt.show()
```
